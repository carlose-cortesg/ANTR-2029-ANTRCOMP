```python
import sqlite3
import pandas as pd

# Conectar a la base de datos SQLite
conn = sqlite3.connect('music_smaller_50000.db')

# Consultar datos, por ejemplo, cargando datos de una tabla
df = pd.read_sql_query("SELECT * FROM playlists", conn)

# Cerrar la conexión

conn.close()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['num_tracks'], bins=30, kde=True)
plt.title('Distribución del Número de Canciones por Playlist')
plt.xlabel('Número de Canciones')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.histplot(df['duration_ms'], bins=30, color='green', kde=True)
plt.title('Distribución de Duración Total de Playlists')
plt.xlabel('Duración Total (ms)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Seleccionar solo algunas columnas para la correlación
columns_of_interest = ['num_tracks', 'num_albums', 'num_followers', 'duration_ms']
sns.pairplot(df[columns_of_interest])
plt.suptitle('Relación de pares y distribución de características de playlists', y=1.02)
plt.show()
```
