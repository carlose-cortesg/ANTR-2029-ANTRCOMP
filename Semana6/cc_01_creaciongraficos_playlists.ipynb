```python
import sqlite3
import pandas as pd

# Conectar a la base de datos SQLite
conn = sqlite3.connect('music_smaller_50000.db')

# Consultar datos, por ejemplo, cargando datos de una tabla
df = pd.read_sql_query("SELECT * FROM playlists", conn)

# Cerrar la conexi√≥n

conn.close()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['num_tracks'], bins=30, kde=True)
plt.title('Distribuci√≥n del N√∫mero de Canciones por Playlist')
plt.xlabel('N√∫mero de Canciones')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.histplot(df['duration_ms'], bins=30, color='green', kde=True)
plt.title('Distribuci√≥n de Duraci√≥n Total de Playlists')
plt.xlabel('Duraci√≥n Total (ms)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Seleccionar solo algunas columnas para la correlaci√≥n
columns_of_interest = ['num_tracks', 'num_albums', 'num_followers', 'duration_ms']
sns.pairplot(df[columns_of_interest])
plt.suptitle('Relaci√≥n de pares y distribuci√≥n de caracter√≠sticas de playlists', y=1.02)
plt.show()
```
