```python
# PREGUNTA 8 Calcular la cantidad total de canciones en cada playlist y ordenarlas de mayor a menor

# CREE UNA TABLA SOBRE LA TABLA DE CANCIONES DONDE AGRUPAMOS POR PID PARA CONTAR EL NUMOERO TOTAL DE CANCIONES ASOCIADAS A UNA PID. LUEGO HICE JOIN DE LA PLAYLIST CON LA TABLA QUE CREE PARA ASOCIAR LA CANTIDAD TOTAL DE CANCIONES. 
%sql WITH grouped_songs AS (SELECT pid, count(*) as num_songs FROM song GROUP BY pid) SELECT p.*, gs.num_songs FROM playlists p JOIN grouped_songs gs ON gs.pid = p.pid ORDER BY gs.num_songs DESC LIMIT 10
 #O # o
%sql SELECT * FROM playlists ORDER BY num_tracks DESC LIMIT 10
```

```python
#PREGUNTA 9 (TERCER BLOQUE) Calcular la duración total y promedio de todas las canciones en una playlist específica
#EXPLICACION: CREE UNA VARAIBLE PARA DESPUES PODER HACER EL JOIN ENTRE LA PLAYLIST ESPECIFICA Y LAS CANCIONES PARA LUEGO SACAR LA SUMATORIA Y EL PROMEDIO DE LA DURACION Y EL PROMEDIO DE LAS CANCIONES DENTRO DE ESA LISTA.
%sql SELECT p.*, SUM(s.duration_ms) as total_duration, AVG(s.duration_ms) as avg_duration FROM playlists p JOIN song s ON s.pid = p.pid WHERE p.pid = 6 GROUP BY p.pid
```

```python
#PREGUNTA 10 (TERCER BLOQUE) Contar la cantidad total de playlists que contienen más de 10 canciones -> Utilizar un COUNT(*)
#HICE UNA TABLA INTERMEDIA EN LA QUE SELECCIONE EL PID Y QUE AGRUPARA LAS CANCIONES EN ESE PID. DESPUES HICE UN JOIN CON PLAYLISTS PARA DECIRLE QUE ME DIERA LAS PLAYLISTS CON CANTIDAD DE CANCIONES MAYOR A DIEZ. 
%sql WITH num_songs_per_pid AS (SELECT pid, COUNT(*) as total_songs FROM song GROUP BY pid) SELECT COUNT(*) FROM playlists p JOIN num_songs_per_pid nspp ON nspp.pid = p.pid WHERE nspp.total_songs > 10
```

```python
#PREGUNTA 11 (TERCER BLOQUE) Obtener el promedio de canciones por playlist
#CREE UNA VARIABLE DE NUMERO DE CANCONES POR PID (NSPP). LE PEDI QUE SELECCIONARA EL NUMERO DE CANCIONES POR PLAYLIS PARA DESPUES DARME EL PROMEDIO DE CANCIONES ASOCIADAS AL TOTAL DE LISTAS. 
%sql WITH num_songs_per_pid AS (SELECT pid, COUNT(*) as total_songs FROM song GROUP BY pid) SELECT AVG(nspp.total_songs) as avg_songs FROM playlists p JOIN num_songs_per_pid nspp ON nspp.pid = p.pid WHERE nspp.total_songs > 10
```

```python
#PREGUNTA 12 (TERCER BLOQUE) Calcular la duración total de todas las canciones en playlists con más de 50 tracks
#USE UN CODIGO ANTERIOR,HICE UN JOIN CON LAS CANCIONES Y FILTRE POR AQUELLAS QUE TUVIERAN MAS DE 50 CANCIONES POR LISTA. 

%sql SELECT p.*, SUM(s.duration_ms) as total_duration FROM playlists p JOIN song s ON s.pid = p.pid WHERE p.num_tracks > 50 GROUP BY p.pid LIMIT 20
```
