# BLOQUE 1


1. Top 20 de Playlists mas Recientes -> Utilizar modified at mayor que

```python
query = """
SELECT *
FROM playlists
ORDER BY modified_at 
LIMIT 20;
"""
```

```python
pd.read_sql_query(query, conn)
```

2. Contar el número total de playlists -> Utilizar COUNT()

```python
query = """
SELECT COUNT(*) AS playlist_count
FROM playlists;
"""
```

```python
pd.read_sql_query(query,conn)
```

9. Nombres de las 10 canciones mas largas en duración

# Bloque 2

4. Contar la cantidad total de playlists que contienen más de 10 canciones -> Utilizar WHERE

```python
query = """ 
SELECT count(*) AS Total_playlists
FROM playlists
WHERE num_tracks > 10


"""
```

```python
pd.read_sql_query(query,conn)
```

5. Encontrar la duración más larga y más corta de todas las canciones

```python
query = """
SELECT 
    MAX(duration_ms) AS Duracion_mas_larga,
    MIN(duration_ms) AS Duracion_mas_corta
FROM song;

"""
```

```python
pd.read_sql_query(query,conn)
```

6. Calcular la suma total de duraciones de todas las canciones de un artista específico

```python
query = """ 
SELECT SUM(duration_ms) AS Suma_duracion_total
FROM song
WHERE artist_name = 'Eminem';
"""
```

```python
pd.read_sql_query(query,conn)
```

9. Número de canciones por lista -> Utilizar CASE

```python
query = """
SELECT pid,SUM(CASE WHEN pid is not null THEN 1 ELSE 0 END) as num_songs
FROM song
GROUP BY pid;

"""
```

```python
pd.read_sql_query(query,conn)
```

11. Seleccionar las canciones que duran más de la duración promedio de todas las canciones

```python
query = """
SELECT *
FROM song
WHERE duration_ms > (SELECT AVG(duration_ms)
FROM song
);

"""
```

```python
pd.read_sql_query(query,conn)
```

# Bloque 3

3. Contar la cantidad de playlists que contienen más de una canción de un álbum específico

```python
query = """ 
SELECT pid, COUNT(DISTINCT album_uri) as num_albums
FROM song
GROUP BY pid
"""
```

```python
pd.read_sql_query(query,conn)
```


5. Obtener el promedio de canciones por playlist

```python
query = """ Select avg(num_tracks) AS Promedio_Canciones_por_Playlist
From playlists;
"""
```

```python
pd.read_sql_query(query,conn)
```

9. Calcular la duración total de todas las canciones en playlists con más de 50 tracks

```python
query = """ 
SELECT 
    SUM (duration_ms) AS Duracion_total
    FROM Playlists
WHERE num_tracks <50
GROUP BY num_tracks
;
"""
```
