```
#Pregunta 6
%%sql
SELECT SUM(duration_ms) AS total_duration
FROM song
WHERE artist_name = 'Daddy Yankee';
```

```
#Pregunta 5
%%sql SELECT
MAX(duration_ms) AS max_duration, MIN(duration_ms) AS min_duration
FROM song;



```

```
#Pregunta 9
%%sql
SELECT
    pid,
    SUM(CASE WHEN pid IS NOT NULL THEN 1 ELSE 0 END) AS song_count
FROM
    song
GROUP BY
    pid
ORDER BY
    pid ASC;
```

```
#Pregunta 2
%%sql SELECT
    SUM(s.duration_ms) AS total_duration_ms
FROM
    Song s
JOIN
    Playlists p ON s.pid = p.pid
WHERE
    s.artist_name = 'Daddy Yankee';






```

```
#Pregunta 4
%%sql SELECT
    p.pid AS playlist_id,
    COUNT(s.track_uri) AS total_songs
FROM
    playlists p
JOIN
    song s ON p.pid = s.pid
GROUP BY
    p.pid
ORDER BY
    total_songs DESC;
```

```
#Pregunta 6
%%sql SELECT COUNT(*) AS total_playlists_with_more_than_10_songs
FROM (
    SELECT
        p.pid
    FROM
        Playlists p
    JOIN
        Song s ON p.pid = s.pid
    GROUP BY
        p.pid
    HAVING
        COUNT(num_tracks) > 10
) AS playlists_with_more_than_10_songs;
```
