## 3 B3 Contar la cantidad de playlists que contienen más de un álbum diferente Que hice: Esta consulta SQL cuenta cuántas listas de reproducción tienen canciones de más de 5 álbumes diferentes. Utiliza una subconsulta que selecciona el pid de la tabla playlists, une esta tabla con song donde los pid coinciden, agrupa los resultados por pid y filtra los grupos que tienen más de 5 álbumes distintos. Luego, la consulta principal cuenta estos grupos y le da el nombre cantidad_playlists_con_multiples_albumes al resultado.

```python
%%sql
SELECT COUNT(*) AS cantidad_playlists_con_multiples_albumes
FROM (
    SELECT pc.pid
    FROM playlists as pc
    JOIN song as c ON pc.pid = c.pid
    GROUP BY pc.pid
    HAVING COUNT(DISTINCT c.album_uri) > 5
) AS playlists_multiples_albumes;
```

## 1 B3 Calcular la duración total y promedio de todas las canciones en una playlist específica Que hice: En essta instrucción SQL suma todas las duraciones (duration_ms) de las canciones y calcula la duración promedio de las mismas, solo para aquellas canciones que pertenecen a la lista de reproducción con pid igual a 6. La consulta da como resultado dos valores: la duracion_total y la duracion_promedio de estas canciones.

```python
%%sql
SELECT 
    SUM(duration_ms) AS duracion_total,
    AVG(duration_ms) AS duracion_promedio
FROM 
    song
WHERE 
    pid = 6;

```
