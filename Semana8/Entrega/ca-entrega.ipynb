## 1 . Histograma / Pregunta:¿Cuál es la distribución de la duración de las pistas en milisegundos y cómo se interpreta la gráfica de frecuencias?

La gráfica muestra un histograma de la distribución de la duración de las pistas en milisegundos (ms). En el eje horizontal (x) se representa el número de pistas agrupadas en intervalos de duración, mientras que en el eje vertical (y) se indica la cantidad de pistas que caen dentro de cada intervalo de duración.

Interpretando la gráfica, observamos que la mayoría de las pistas tienen una duración relativamente corta, concentrándose principalmente en los primeros intervalos a la izquierda del histograma. La frecuencia de pistas disminuye conforme la duración aumenta, lo cual sugiere que hay pocas pistas con duraciones largas. Esto puede indicar que en el conjunto de datos, es más común encontrar pistas de corta duración que de larga duración.

```python
plt.figure(figsize=(10, 6))
plt.hist(x=variable,bins=20)
plt.title('Número de Pistas vs Duración')
plt.xlabel('Número de Pistas')
plt.ylabel('Duración (ms)')
plt.show()
```

```python
import sqlite3
import pandas as pd

# Conectarse a la base de datos
connection = sqlite3.connect('music_smaller_10000.db')

# Crear el query SQL como una cadena y ejecutarlo para transformar en un DataFrame
query = """
SELECT
    pid,
    num_followers,
    CASE
        WHEN num_followers < 100 THEN 'Baja'
        WHEN num_followers >= 100 AND num_followers <= 1000 THEN 'Media'
        ELSE 'Alta'
    END AS popularidad
FROM playlists;
"""
canciones = pd.read_sql_query(query, connection)

# Mostrar cómo queda el DataFrame
print(canciones.head())

# Cerrar la conexión
connection.close()
```

## Grafico Categorico / ¿Cuál es la distribución del número de seguidores (num_followers) en cada categoría de popularidad (Baja, Media, Alta) y qué conclusiones se pueden extraer sobre la popularidad de las playlists basándose en esta distribución?
La gráfica muestra un boxplot de la distribución del número de seguidores (num_followers) categorizados en niveles de popularidad: Baja, Media y Alta, aunque solo se representan las categorías Baja y Alta. En la categoría Baja, la mediana está cerca de 0, con una caja (IQR) estrecha y algunos valores atípicos ligeramente superiores a 0. En la categoría Alta, la mediana es mucho mayor (alrededor de 4000) y el IQR muestra una gran variabilidad en el número de seguidores, extendiéndose hasta aproximadamente 16000 sin valores atípicos prominentes. Las playlists de baja popularidad tienen un alcance muy limitado y consistente, mientras que las de alta popularidad muestran una amplia variabilidad y un gran número de seguidores. La ausencia de datos en la categoría Media sugiere un posible error de clasificación o falta de datos en esa categoría. Esta información puede ayudar a los creadores de contenido a desarrollar estrategias para aumentar la popularidad de sus playlists, moviéndolas de la categoría Baja a la Alta.

```python
# Crear un boxplot utilizando Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='popularidad', y='num_followers', data=canciones)

# Añadir título y etiquetas
plt.title('Boxplot de num_followers por categoría de popularidad')
plt.xlabel('Popularidad')
plt.ylabel('Número de seguidores')

# Mostrar el gráfico
plt.show()
```
