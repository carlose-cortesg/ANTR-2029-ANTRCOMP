**INICIO DESARROLLO DE GRFICOS**

**GRFICO 1:**
Este grfico muestra que, a medida que aumenta el nmero de canciones en una lista de reproduccin, tambin incrementa su duracin en milisegundos. Por lo tanto, podemos concluir que estas variables son directamente proporcionales.

```python
sns.relplot(data=playlists, x="num_tracks", y="duration_ms")
plt.xlabel ("Nmero de canciones")
plt.ylabel ("Duracin en milisegundos")
```

**GRFICO 2:**
Este grfico, al igual que el anterior, nos indica que mientras ms canciones tiene una lista de reproduccin, mayor va a ser su duracin en milisegundos, la diferencia, es que se ha aadido una variable cualitativa, la cual es num_artist. Esta, en relacin a la grfica, nos indica que mientras ms canciones tenga una lista de reproduccin, habr consecuentemente ms artistas.

```python
sns.relplot(playlists, x="num_tracks", y="duration_ms", hue= "num_artists", kind= "scatter")
```

**GRFICO 3:**
El grfico a continuacin ilustra cmo se distribuye la cantidad de canciones en las listas de reproduccin.
```python
sns.histplot (playlists, x= "num_tracks")
plt.xlabel ("Numero de canciones")
plt.ylabel ("Frecuencia")
```
