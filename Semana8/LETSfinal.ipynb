**INICIO DESARROLLO DE GRÁFICOS**

**GRÁFICO 1:**
Este gráfico muestra que, a medida que aumenta el número de canciones en una lista de reproducción, también incrementa su duración en milisegundos. Por lo tanto, podemos concluir que estas variables son directamente proporcionales.

```python
sns.relplot(data=playlists, x="num_tracks", y="duration_ms")
plt.xlabel ("Número de canciones")
plt.ylabel ("Duración en milisegundos")
```

**GRÁFICO 2:**
Este gráfico, al igual que el anterior, nos indica que mientras más canciones tiene una lista de reproducción, mayor va a ser su duración en milisegundos, la diferencia, es que se ha añadido una variable cualitativa, la cual es num_artist. Esta, en relación a la gráfica, nos indica que mientras más canciones tenga una lista de reproducción, habrá consecuentemente más artistas.

```python
sns.relplot(playlists, x="num_tracks", y="duration_ms", hue= "num_artists", kind= "scatter")
```

**GRÁFICO 3:**
El gráfico a continuación ilustra cómo se distribuye la cantidad de canciones en las listas de reproducción. El eje x muestra los rangos de la variable numérica "num_tracks", mientras que el eje y indica la frecuencia de datos en cada uno de estos rangos. Así, la distribución de los intervalos revela que la mayoría de las listas de reproducción contienen pocas canciones, y, por ende, hay pocas listas de reproducción con un gran número de canciones. Podríamos decir que la relación entre las variables es inversamente proporcional.

```python
sns.histplot (playlists, x= "num_tracks")
plt.xlabel ("Numero de canciones")
plt.ylabel ("Frecuencia")
```

**GRÁFICO 4:**
Este gráfico nos está enseñando la relación que hay entre cada una de las canciones de Misfits y su duración en milisegundos a la vez que los clasifica, según el color de su barra, a su álbum correspondiente. De esta, sería posible calcular manualmente el tiempo promedio de duración de las canciones de cada álbum; o también se podría cambiar la línea de código 2, cambiando "track_name" por "album_name" y así se obtendría la duración promedio del álbum.

```python
misfits_songs = song[song['artist_name'] == 'Misfits']
sns.catplot(misfits_songs, x="track_name", y= "duration_ms", kind= "bar", hue ="album_name")
plt.xlabel("Nombre del álbum")
plt.ylabel("Duración en ms")
plt.xticks(rotation=90, ha='right')
```
